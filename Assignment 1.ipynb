```python
import pandas as pd
import numpy as np
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nifty_stocks=["RELIANCE.NS","HDFCBANK.NS","TCS.NS","ICICIBANK.NS","HINDUNILVR.NS",
              "INFY.NS","ITC.NS","BHARTIARTL.NS","BAJFINANCE.NS","SBIN.NS","LT.NS","KOTAKBANK.NS","HCLTECH.NS",
             "MARUTI.NS","AXISBANK.NS","TITAN.NS","SUNPHARMA.NS","ADANIENT.NS","ASIANPAINT.NS","NESTLEIND.NS"]
#download data
data=yf.download(nifty_stocks, start="2022-01-01",end="2023-01-01")

#calculaing returns
returns=data['Adj Close'].pct_change()

#Correlation Matrix
corr_matrix=returns.corr()

#Avg correlation
avg_corr_matrix=corr_matrix.mean()

#Sorting the avg corr matrix and selecting the first 5(least correlated)
sorted_avg_corr_matrix=avg_corr_matrix.sort_values()
selected_stocks=sorted_avg_corr_matrix.index[:5]
stock_output='\n'.join(selected_stocks)
print(stock_output)
```

```python
#plotting the matrix
selected_corr_matrix = corr_matrix.loc[selected_stocks, selected_stocks]
plt.figure(figsize=(5,5))
sns.heatmap(selected_corr_matrix,annot=True,cmap='coolwarm',fmt="0.2f")
plt.title("Correlation Matrix")
plt.show()
```
